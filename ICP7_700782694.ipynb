```python
#LeNet and AlexNet

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Define Models (ARCHITECTURE UNCHANGED)
def build_lenet():
    model = keras.models.Sequential([
        keras.layers.Conv2D(6, kernel_size=5, strides=1,
                            activation='relu',
                            input_shape=(32,32,3),
                            padding='same'),
        keras.layers.AveragePooling2D(pool_size=(2,2)),
        keras.layers.Conv2D(16, kernel_size=5, strides=1,
                            activation='relu',
                            padding='valid'),
        keras.layers.AveragePooling2D(pool_size=(2,2)),
        keras.layers.Conv2D(120, kernel_size=5, strides=1,
                            activation='relu',
                            padding='valid'),
        keras.layers.Flatten(),
        keras.layers.Dense(84, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    return model


def build_alexnet():
    model = keras.models.Sequential()
    model.add(keras.layers.Conv2D(16,(3,3),strides=(4,4),
                                  input_shape=(32,32,3),
                                  activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2,2),strides=(2,2)))

    model.add(keras.layers.Conv2D(60,(5,5),padding='same',activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2,2),strides=(2,2)))

    model.add(keras.layers.Conv2D(60,(3,3),padding='same',activation='relu'))
    model.add(keras.layers.Conv2D(30,(3,3),padding='same',activation='relu'))
    model.add(keras.layers.Conv2D(20,(3,3),padding='same',activation='relu'))

    model.add(keras.layers.MaxPooling2D(pool_size=(2,2),strides=(2,2)))
    model.add(keras.layers.Flatten())

    model.add(keras.layers.Dense(200, activation='relu'))
    model.add(keras.layers.Dropout(0.1))
    model.add(keras.layers.Dense(200, activation='relu'))
    model.add(keras.layers.Dropout(0.1))
    model.add(keras.layers.Dense(10,activation='softmax'))
    return model


# 2. Helper Functions
def preprocess_to_32x32x3(x):
    if x.ndim == 3:
        x = x[..., np.newaxis]
    x = tf.image.resize(x, (32,32))
    x = tf.image.grayscale_to_rgb(x)
    return x.numpy()/255.0

def plot_subplots(history, title):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(title + " Accuracy")
    plt.legend(["Train","Val"])

    plt.subplot(1,2,2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(title + " Loss")
    plt.legend(["Train","Val"])
    plt.show()

def plot_conf_matrix(y_true,y_pred,title):
    cm = confusion_matrix(y_true,y_pred)
    plt.figure(figsize=(6,5))
    plt.imshow(cm)
    plt.title(title)
    plt.colorbar()
    plt.show()

def plot_class_distribution(y,title):
    counts=np.bincount(y)
    plt.figure(figsize=(6,4))
    plt.bar(range(len(counts)),counts)
    plt.title(title)
    plt.show()

def show_misclassified(x,y_true,y_pred,title):
    idx=np.where(y_true!=y_pred)[0][:16]
    plt.figure(figsize=(8,8))
    plt.suptitle(title)
    for i,id in enumerate(idx):
        plt.subplot(4,4,i+1)
        plt.imshow(x[id])
        plt.axis("off")
    plt.show()


# 3. Training Pipeline (Hyperparameter Tuning Added)
def train_model(model,x_train,y_train,x_test,y_test,
                optimizer,model_name,dataset_name,improved=False):

    model.compile(optimizer=optimizer,
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])

    callbacks=[]
    if improved:
        callbacks=[
            keras.callbacks.ReduceLROnPlateau(patience=3,factor=0.5),
            keras.callbacks.EarlyStopping(patience=6,restore_best_weights=True),
            keras.callbacks.ModelCheckpoint(f"{dataset_name}_{model_name}_best.keras",
                                            save_best_only=True)
        ]

    history=model.fit(x_train,y_train,
                      epochs=100,
                      batch_size=64,
                      validation_data=(x_test,y_test),
                      callbacks=callbacks,
                      verbose=1)

    loss,acc=model.evaluate(x_test,y_test,verbose=0)
    print(f"{dataset_name}-{model_name} Test Accuracy:",acc)

    y_pred=np.argmax(model.predict(x_test),axis=1)
    return history,y_pred

# 4. DATASET 1 – CIFAR10
(x_train_c,y_train_c),(x_test_c,y_test_c)=keras.datasets.cifar10.load_data()
y_train_c=y_train_c.reshape(-1)
y_test_c=y_test_c.reshape(-1)

x_train_c=x_train_c/255.0
x_test_c=x_test_c/255.0

plot_class_distribution(y_train_c,"CIFAR10 Class Distribution")

# Baseline LeNet
hist_base,pred_base=train_model(build_lenet(),
                                x_train_c,y_train_c,
                                x_test_c,y_test_c,
                                keras.optimizers.Adam(),
                                "LeNet","CIFAR10",False)

plot_subplots(hist_base,"CIFAR10 LeNet Baseline")
plot_conf_matrix(y_test_c,pred_base,"CIFAR10 LeNet Baseline CM")

# Improved LeNet
hist_imp,pred_imp=train_model(build_lenet(),
                              x_train_c,y_train_c,
                              x_test_c,y_test_c,
                              keras.optimizers.Adam(learning_rate=0.001),
                              "LeNet","CIFAR10",True)

plot_subplots(hist_imp,"CIFAR10 LeNet Improved")
plot_conf_matrix(y_test_c,pred_imp,"CIFAR10 LeNet Improved CM")
show_misclassified(x_test_c,y_test_c,pred_imp,"Misclassified CIFAR10")


# 5. DATASET 2 – Fashion MNIST
(x_train_f,y_train_f),(x_test_f,y_test_f)=keras.datasets.fashion_mnist.load_data()

x_train_f=preprocess_to_32x32x3(x_train_f)
x_test_f=preprocess_to_32x32x3(x_test_f)

plot_class_distribution(y_train_f,"Fashion MNIST Distribution")

# Baseline AlexNet
hist_base_f,pred_base_f=train_model(build_alexnet(),
                                    x_train_f,y_train_f,
                                    x_test_f,y_test_f,
                                    keras.optimizers.SGD(),
                                    "AlexNet","FashionMNIST",False)

plot_subplots(hist_base_f,"FashionMNIST AlexNet Baseline")
plot_conf_matrix(y_test_f,pred_base_f,"FashionMNIST Baseline CM")

# Improved AlexNet
hist_imp_f,pred_imp_f=train_model(build_alexnet(),
                                  x_train_f,y_train_f,
                                  x_test_f,y_test_f,
                                  keras.optimizers.SGD(learning_rate=0.01,momentum=0.9),
                                  "AlexNet","FashionMNIST",True)

plot_subplots(hist_imp_f,"FashionMNIST AlexNet Improved")
plot_conf_matrix(y_test_f,pred_imp_f,"FashionMNIST Improved CM")
show_misclassified(x_test_f,y_test_f,pred_imp_f,"Misclassified FashionMNIST")
```

```python
#VGG-19

# VGG19 Baseline + Improved Model (CIFAR-10)
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG19
from tensorflow.keras.applications.vgg19 import preprocess_input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import confusion_matrix, classification_report


# Load CIFAR-10 Dataset
(x_train_full, y_train_full), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

num_classes = 10
class_names = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]

# Train / Validation Split
val_size = int(0.2 * len(x_train_full))
x_val = x_train_full[:val_size]
y_val = y_train_full[:val_size]
x_train = x_train_full[val_size:]
y_train = y_train_full[val_size:]

y_train = to_categorical(y_train, num_classes)
y_val   = to_categorical(y_val, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

# Data Augmentation
augment = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.1)
])

def preprocess_train(x, y):
    x = tf.cast(x, tf.float32)
    x = augment(x)
    x = preprocess_input(x)
    return x, y

def preprocess_eval(x, y):
    x = tf.cast(x, tf.float32)
    x = preprocess_input(x)
    return x, y

train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(20000).map(preprocess_train).batch(64)
val_ds   = tf.data.Dataset.from_tensor_slices((x_val, y_val)).map(preprocess_eval).batch(64)
test_ds  = tf.data.Dataset.from_tensor_slices((x_test, y_test_cat)).map(preprocess_eval).batch(64)


# BASELINE MODEL
def build_baseline():
    base = VGG19(weights="imagenet", include_top=False, input_shape=(32,32,3))
    base.trainable = False

    inputs = layers.Input(shape=(32,32,3))
    x = base(inputs, training=False)
    x = layers.Flatten()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    return models.Model(inputs, outputs)

baseline = build_baseline()
baseline.compile(optimizer="rmsprop",
                 loss="categorical_crossentropy",
                 metrics=["accuracy"])

baseline.fit(train_ds,
             validation_data=val_ds,
             epochs=10,
             callbacks=[EarlyStopping(patience=3, restore_best_weights=True)],
             verbose=1)

baseline.evaluate(test_ds, verbose=1)

# IMPROVED MODEL
def build_improved():
    base = VGG19(weights="imagenet", include_top=False, input_shape=(32,32,3))
    base.trainable = False

    inputs = layers.Input(shape=(32,32,3))
    x = base(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation="relu")(x)
    x = layers.Dropout(0.4)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    return models.Model(inputs, outputs)

improved = build_improved()
improved.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                 loss="categorical_crossentropy",
                 metrics=["accuracy"])

callbacks = [
    ModelCheckpoint("best_model.keras", save_best_only=True, monitor="val_accuracy"),
    EarlyStopping(patience=5, restore_best_weights=True),
    ReduceLROnPlateau(patience=3)
]

history = improved.fit(train_ds,
                       validation_data=val_ds,
                       epochs=20,
                       callbacks=callbacks,
                       verbose=1)

improved.evaluate(test_ds, verbose=1)

#Save and Load Matrix
model_loaded = tf.keras.models.load_model("best_model.keras")

# Predictions
y_pred_probs = model_loaded.predict(test_ds)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = y_test.flatten()

# CONFUSION MATRIX
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
plt.imshow(cm)
plt.colorbar()
plt.xticks(range(num_classes), class_names, rotation=45)
plt.yticks(range(num_classes), class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# TRAINING & VALIDATION PLOTS
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.legend()
plt.title("Accuracy")

plt.subplot(1,2,2)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Loss")

plt.tight_layout()
plt.show()

# EXTRA VISUALIZATION 1: Sample Predictions
plt.figure(figsize=(10,8))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(x_test[i])
    plt.title(class_names[y_pred[i]])
    plt.axis("off")
plt.tight_layout()
plt.show()


# EXTRA VISUALIZATION 2: Class Distribution
plt.figure()
counts = np.bincount(y_train_full.flatten(), minlength=num_classes)
plt.bar(range(num_classes), counts)
plt.xticks(range(num_classes), class_names, rotation=45)
plt.title("Class Distribution")
plt.tight_layout()
plt.show()

# LOGICAL DESCRIPTIONS
print("\nBaseline vs Improved:")
print("- GlobalAveragePooling reduces parameters and improves generalization.")
print("- Adam optimizer improves convergence.")
print("- Data augmentation reduces overfitting.")
print("- Learning rate scheduling stabilizes training.")
print("- Fine architectural adjustments improved validation behavior.")
```

```python
#VGG- 16


# VGG-Style CNN: Baseline vs Improved (CIFAR-10) + New Dataset
import os, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from sklearn.metrics import confusion_matrix

print("TensorFlow:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))


# Reproducibility
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)


# Utility: plot history
def plot_train_test_curves(history, title_prefix="Model"):
    # Task 9: Training and testing (validation) loss & accuracy in one figure using subplots
    hist = history.history
    fig = plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(hist["loss"], label="train_loss")
    plt.plot(hist.get("val_loss", []), label="val_loss")
    plt.title(f"{title_prefix} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(hist["accuracy"], label="train_acc")
    plt.plot(hist.get("val_accuracy", []), label="val_acc")
    plt.title(f"{title_prefix} - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(cm, class_names, title="Confusion Matrix"):
    # Task 8: confusion matrix plot
    fig = plt.figure(figsize=(8,7))
    plt.imshow(cm, interpolation='nearest')
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45, ha="right")
    plt.yticks(tick_marks, class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

def show_misclassified_images(x, y_true, y_pred, class_names, max_show=16, title="Misclassified Examples"):
    # Extra visualization (Task 3/10): show misclassified examples
    wrong = np.where(y_true != y_pred)[0]
    if len(wrong) == 0:
        print("No misclassifications found ")
        return
    n = min(max_show, len(wrong))
    idx = wrong[:n]

    plt.figure(figsize=(10,10))
    for i, k in enumerate(idx):
        plt.subplot(int(np.ceil(np.sqrt(n))), int(np.ceil(np.sqrt(n))), i+1)
        plt.imshow(x[k])
        plt.title(f"T:{class_names[y_true[k]]}\nP:{class_names[y_pred[k]]}", fontsize=9)
        plt.axis("off")
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

def class_accuracy_bar(y_true, y_pred, class_names, title="Per-class Accuracy"):
    # Extra visualization (Task 3/10): per-class accuracy bar chart
    accs = []
    for c in range(len(class_names)):
        idx = np.where(y_true == c)[0]
        acc = (y_pred[idx] == c).mean() if len(idx) else 0.0
        accs.append(acc)

    plt.figure(figsize=(10,4))
    plt.bar(class_names, accs)
    plt.xticks(rotation=45, ha="right")
    plt.ylim(0, 1)
    plt.title(title)
    plt.ylabel("Accuracy")
    plt.tight_layout()
    plt.show()


# VGG-style blocks
def vgg_block(x, filters, convs=2, use_bn=False, l2=0.0, dropout=0.0):
    reg = regularizers.l2(l2) if l2 > 0 else None
    for _ in range(convs):
        x = layers.Conv2D(filters, (3,3), padding="same", kernel_regularizer=reg)(x)
        if use_bn:
            x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
    x = layers.MaxPooling2D((2,2))(x)
    if dropout > 0:
        x = layers.Dropout(dropout)(x)
    return x

def build_baseline_vgg(input_shape, num_classes):
    # Baseline model: closer to your original style (no BN, no augmentation, bigger head)
    inputs = keras.Input(shape=input_shape)

    x = vgg_block(inputs, 64, convs=2, use_bn=False)
    x = vgg_block(x, 128, convs=2, use_bn=False)
    x = vgg_block(x, 256, convs=3, use_bn=False)
    x = vgg_block(x, 512, convs=3, use_bn=False)

    x = layers.Flatten()(x)
    x = layers.Dense(1024, activation="relu")(x)
    x = layers.Dropout(0.4)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return keras.Model(inputs, outputs, name="Baseline_VGG")

def build_improved_vgg(input_shape, num_classes, aug=True):
    # Improved model: augmentation + BN + L2 + dropout + GAP head (fewer params, better generalization)
    inputs = keras.Input(shape=input_shape)

    x = inputs
    if aug:
        data_aug = keras.Sequential([
            layers.RandomFlip("horizontal"),
            layers.RandomTranslation(0.1, 0.1),
            layers.RandomRotation(0.1),
            layers.RandomZoom(0.1),
        ], name="data_augmentation")
        x = data_aug(x)

    x = vgg_block(x, 64,  convs=2, use_bn=True, l2=1e-4, dropout=0.05)
    x = vgg_block(x, 128, convs=2, use_bn=True, l2=1e-4, dropout=0.10)
    x = vgg_block(x, 256, convs=3, use_bn=True, l2=1e-4, dropout=0.15)
    x = vgg_block(x, 512, convs=3, use_bn=True, l2=1e-4, dropout=0.20)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.4)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return keras.Model(inputs, outputs, name="Improved_VGG")


# Train/Evaluate pipeline
def train_and_evaluate(model, x_train, y_train, x_test, y_test, class_names, title, target_acc=0.80):
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=3e-4),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    callbacks = [
        keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=8, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6, verbose=1),
    ]

    history = model.fit(
        x_train, y_train,
        validation_split=0.15,
        epochs=60,
        batch_size=128,
        callbacks=callbacks,
        verbose=1
    )

    # Task 6: Evaluate on testing set
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"\n {title} Test Accuracy: {test_acc*100:.2f}% | Test Loss: {test_loss:.4f}")

    if target_acc is not None:
        print(f"Target accuracy: > {target_acc*100:.0f}%")

    # Predictions
    probs = model.predict(x_test, verbose=0)
    y_pred = np.argmax(probs, axis=1)

    # Task 8: confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Task 9: loss/accuracy plots using history + subplots
    plot_train_test_curves(history, title_prefix=title)

    # Task 8: confusion matrix plot
    plot_confusion_matrix(cm, class_names, title=f"{title} - Confusion Matrix")

    # Task 3/10: additional visualizations
    show_misclassified_images(x_test, y_test, y_pred, class_names, max_show=16, title=f"{title} - Misclassified")
    class_accuracy_bar(y_test, y_pred, class_names, title=f"{title} - Per-class Accuracy")

    return history, (test_loss, test_acc), (y_pred, probs), cm


# PART A: CIFAR-10 (Baseline + Improved)
# Task 4: Use dataset (CIFAR-10 baseline provided)
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
y_train = y_train.squeeze()
y_test = y_test.squeeze()

# Normalize
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

cifar_classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]

# Visualization from your original style (sample images) - plus extra (Task 3/10)
plt.figure(figsize=(8,3))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i])
    plt.title(cifar_classes[y_train[i]], fontsize=9)
    plt.axis("off")
plt.suptitle("CIFAR-10: Sample Training Images")
plt.tight_layout()
plt.show()

# Baseline
baseline_cifar = build_baseline_vgg(input_shape=x_train.shape[1:], num_classes=10)
baseline_cifar.summary()

hist_base_cifar, eval_base_cifar, pred_base_cifar, cm_base_cifar = train_and_evaluate(
    baseline_cifar, x_train, y_train, x_test, y_test, cifar_classes,
    title="CIFAR-10 Baseline VGG", target_acc=None
)

# Improved (target >80%)
improved_cifar = build_improved_vgg(input_shape=x_train.shape[1:], num_classes=10, aug=True)
improved_cifar.summary()

hist_imp_cifar, eval_imp_cifar, pred_imp_cifar, cm_imp_cifar = train_and_evaluate(
    improved_cifar, x_train, y_train, x_test, y_test, cifar_classes,
    title="CIFAR-10 Improved VGG", target_acc=0.80
)

# Task 7: Save improved model and use for prediction
save_path_cifar = "improved_vgg_cifar10.keras"
improved_cifar.save(save_path_cifar)
print("\n  Saved improved CIFAR-10 model to:", save_path_cifar)

loaded_cifar = keras.models.load_model(save_path_cifar)
sample_idx = 1000
sample = x_test[sample_idx:sample_idx+1]
proba = loaded_cifar.predict(sample, verbose=0)[0]
pred_class = np.argmax(proba)

plt.figure(figsize=(3,3))
plt.imshow(x_test[sample_idx])
plt.title(f"Loaded Model Prediction: {cifar_classes[pred_class]}\nTrue: {cifar_classes[y_test[sample_idx]]}")
plt.axis("off")
plt.show()


# PART B: New Dataset (Fashion-MNIST) Baseline + Improved

# Task 4: Use dataset of your choice (Fashion-MNIST)
(fx_train, fy_train), (fx_test, fy_test) = keras.datasets.fashion_mnist.load_data()

# Normalize and convert to 3-channel 32x32 to reuse VGG-style model
fx_train = fx_train.astype("float32") / 255.0
fx_test  = fx_test.astype("float32") / 255.0

fx_train = np.expand_dims(fx_train, -1)
fx_test  = np.expand_dims(fx_test, -1)

# Resize to 32x32 and convert to 3 channels
fx_train = tf.image.resize(fx_train, (32,32)).numpy()
fx_test  = tf.image.resize(fx_test,  (32,32)).numpy()

fx_train = np.repeat(fx_train, 3, axis=-1)
fx_test  = np.repeat(fx_test,  3, axis=-1)

fashion_classes = [
    "T-shirt/top","Trouser","Pullover","Dress","Coat",
    "Sandal","Shirt","Sneaker","Bag","Ankle boot"
]

# Extra visualization (Task 3/10)
plt.figure(figsize=(8,3))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(fx_train[i])
    plt.title(fashion_classes[fy_train[i]], fontsize=9)
    plt.axis("off")
plt.suptitle("Fashion-MNIST: Sample Training Images (Resized to 32x32x3)")
plt.tight_layout()
plt.show()

# Task 4: implement baseline model on new dataset
baseline_fashion = build_baseline_vgg(input_shape=fx_train.shape[1:], num_classes=10)
baseline_fashion.summary()

hist_base_fash, eval_base_fash, pred_base_fash, cm_base_fash = train_and_evaluate(
    baseline_fashion, fx_train, fy_train, fx_test, fy_test, fashion_classes,
    title="Fashion-MNIST Baseline VGG", target_acc=None
)

# Task 5: apply modified architecture on new dataset
# Slightly different augmentation (no horizontal flip needed for fashion; use small rotations/translations)
def build_improved_vgg_fashion(input_shape, num_classes):
    inputs = keras.Input(shape=input_shape)
    data_aug = keras.Sequential([
        layers.RandomTranslation(0.08, 0.08),
        layers.RandomRotation(0.08),
        layers.RandomZoom(0.08),
    ], name="data_augmentation_fashion")
    x = data_aug(inputs)

    x = vgg_block(x, 64,  convs=2, use_bn=True, l2=1e-4, dropout=0.05)
    x = vgg_block(x, 128, convs=2, use_bn=True, l2=1e-4, dropout=0.10)
    x = vgg_block(x, 256, convs=3, use_bn=True, l2=1e-4, dropout=0.15)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.35)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    return keras.Model(inputs, outputs, name="Improved_VGG_Fashion")

improved_fashion = build_improved_vgg_fashion(input_shape=fx_train.shape[1:], num_classes=10)
improved_fashion.summary()

hist_imp_fash, eval_imp_fash, pred_imp_fash, cm_imp_fash = train_and_evaluate(
    improved_fashion, fx_train, fy_train, fx_test, fy_test, fashion_classes,
    title="Fashion-MNIST Improved VGG", target_acc=None
)

# Task 7: save improved model and predict
save_path_fashion = "improved_vgg_fashionmnist.keras"
improved_fashion.save(save_path_fashion)
print("\n Saved improved Fashion-MNIST model to:", save_path_fashion)

loaded_fashion = keras.models.load_model(save_path_fashion)
sample_idx = 123
sample = fx_test[sample_idx:sample_idx+1]
proba = loaded_fashion.predict(sample, verbose=0)[0]
pred_class = np.argmax(proba)

plt.figure(figsize=(3,3))
plt.imshow(fx_test[sample_idx])
plt.title(f"Loaded Model Prediction: {fashion_classes[pred_class]}\nTrue: {fashion_classes[fy_test[sample_idx]]}")
plt.axis("off")
plt.show()
```

```python

# VGG-16 Style CNN
# Baseline vs Improved
# CIFAR-10 + Fashion-MNIST
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from sklearn.metrics import confusion_matrix


# Configuration
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

EPOCHS = 60
BATCH_SIZE = 128
LR = 3e-4

print("TensorFlow:", tf.__version__)
print("GPU:", tf.config.list_physical_devices("GPU"))


# Utility Functions
def plot_history(history, title):
    fig, ax = plt.subplots(1, 2, figsize=(12,5))

    ax[0].plot(history.history['loss'], label='Train')
    ax[0].plot(history.history['val_loss'], label='Validation')
    ax[0].set_title(f"{title} - Loss")
    ax[0].legend()

    ax[1].plot(history.history['accuracy'], label='Train')
    ax[1].plot(history.history['val_accuracy'], label='Validation')
    ax[1].set_title(f"{title} - Accuracy")
    ax[1].legend()

    plt.show()


def plot_conf_matrix(y_true, y_pred, classes, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8,6))
    plt.imshow(cm)
    plt.title(title)
    plt.colorbar()
    plt.xticks(range(len(classes)), classes, rotation=45)
    plt.yticks(range(len(classes)), classes)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()


def show_misclassified(x, y_true, y_pred, classes, n=16):
    wrong = np.where(y_true != y_pred)[0][:n]

    plt.figure(figsize=(8,8))
    for i, idx in enumerate(wrong):
        plt.subplot(4,4,i+1)
        plt.imshow(x[idx])
        plt.title(f"T:{classes[y_true[idx]]}\nP:{classes[y_pred[idx]]}")
        plt.axis("off")
    plt.tight_layout()
    plt.show()



# VGG Block
def vgg_block(x, filters, conv_layers=2, batch_norm=False, l2_reg=0, dropout=0):
    reg = regularizers.l2(l2_reg) if l2_reg > 0 else None

    for _ in range(conv_layers):
        x = layers.Conv2D(filters, (3,3), padding="same",
                          kernel_regularizer=reg)(x)
        if batch_norm:
            x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)

    x = layers.MaxPooling2D((2,2))(x)

    if dropout > 0:
        x = layers.Dropout(dropout)(x)

    return x



# Model Builders
def build_baseline(input_shape, num_classes):
    inputs = keras.Input(shape=input_shape)

    x = vgg_block(inputs, 64)
    x = vgg_block(x, 128)
    x = vgg_block(x, 256, conv_layers=3)
    x = vgg_block(x, 512, conv_layers=3)

    x = layers.Flatten()(x)
    x = layers.Dense(1024, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return keras.Model(inputs, outputs, name="Baseline_VGG")


def build_improved(input_shape, num_classes, augment=True):
    inputs = keras.Input(shape=input_shape)

    x = inputs

    if augment:
        augmentation = keras.Sequential([
            layers.RandomFlip("horizontal"),
            layers.RandomRotation(0.1),
            layers.RandomZoom(0.1),
            layers.RandomTranslation(0.1, 0.1)
        ])
        x = augmentation(x)

    x = vgg_block(x, 64, batch_norm=True, l2_reg=1e-4, dropout=0.05)
    x = vgg_block(x, 128, batch_norm=True, l2_reg=1e-4, dropout=0.1)
    x = vgg_block(x, 256, conv_layers=3, batch_norm=True, l2_reg=1e-4, dropout=0.15)
    x = vgg_block(x, 512, conv_layers=3, batch_norm=True, l2_reg=1e-4, dropout=0.2)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(512, kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return keras.Model(inputs, outputs, name="Improved_VGG")



# Training Function
def train_model(model, x_train, y_train, x_test, y_test, classes, title):

    model.compile(
        optimizer=keras.optimizers.Adam(LR),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    callbacks = [
        keras.callbacks.EarlyStopping(patience=8, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(patience=3)
    ]

    history = model.fit(
        x_train, y_train,
        validation_split=0.15,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=callbacks
    )

    loss, acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"\n{title} Test Accuracy: {acc*100:.2f}%")

    predictions = model.predict(x_test)
    y_pred = np.argmax(predictions, axis=1)

    plot_history(history, title)
    plot_conf_matrix(y_test, y_pred, classes, title)
    show_misclassified(x_test, y_test, y_pred, classes)

    return model



# PART 1 — CIFAR-10
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
y_train = y_train.squeeze()
y_test = y_test.squeeze()

x_train = x_train / 255.0
x_test = x_test / 255.0

cifar_classes = [
    "airplane","automobile","bird","cat","deer",
    "dog","frog","horse","ship","truck"
]

# Baseline
baseline = build_baseline(x_train.shape[1:], 10)
baseline = train_model(baseline, x_train, y_train, x_test, y_test,
                       cifar_classes, "CIFAR-10 Baseline")

# Improved
improved = build_improved(x_train.shape[1:], 10)
improved = train_model(improved, x_train, y_train, x_test, y_test,
                       cifar_classes, "CIFAR-10 Improved")

improved.save("improved_cifar10.keras")



# PART 2 — Fashion-MNIST
(fx_train, fy_train), (fx_test, fy_test) = keras.datasets.fashion_mnist.load_data()

fx_train = fx_train / 255.0
fx_test = fx_test / 255.0

fx_train = np.expand_dims(fx_train, -1)
fx_test = np.expand_dims(fx_test, -1)

fx_train = tf.image.resize(fx_train, (32,32))
fx_test = tf.image.resize(fx_test, (32,32))

fx_train = np.repeat(fx_train, 3, axis=-1)
fx_test = np.repeat(fx_test, 3, axis=-1)

fashion_classes = [
    "T-shirt","Trouser","Pullover","Dress","Coat",
    "Sandal","Shirt","Sneaker","Bag","Boot"
]

# Baseline
baseline_f = build_baseline(fx_train.shape[1:], 10)
baseline_f = train_model(baseline_f, fx_train, fy_train,
                         fx_test, fy_test,
                         fashion_classes, "Fashion Baseline")

# Improved
improved_f = build_improved(fx_train.shape[1:], 10, augment=False)
improved_f = train_model(improved_f, fx_train, fy_train,
                         fx_test, fy_test,
                         fashion_classes, "Fashion Improved")

improved_f.save("improved_fashion.keras")
```

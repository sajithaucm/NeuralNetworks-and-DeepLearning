```python
#Trail 1
#importing set of libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
#importing data from DataSet
#importing the given dataset glass.csv
dst_Data = pd.read_csv("glass.csv")
dst_Data.info()
```

```python
# Last column is Type
X = dst_Data.iloc[:, :-1]
y = dst_Data.iloc[:, -1]

# FIX DATASET: convert to binary classes
# Type 1 and 2 = Window glass (1)
# Others = Non-window glass (0)
y_binary = y.apply(lambda v: 1 if v in [1, 2] else 0)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_binary, test_size=0.2, random_state=0, stratify=y_binary
)

# Gaussian Naive Bayes
gn = GaussianNB()
gn.fit(X_train, y_train)

y_pred = gn.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```

```python
#Trail 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Load the dataset
# If you haven't uploaded it yet, you can use:
# from google.colab import files
# uploaded = files.upload()

df = pd.read_csv('glass.csv')
print("Dataset loaded. Shape:", df.shape)

# Drop duplicates if any
df = df.drop_duplicates()

# Prepare features and target
X = df.drop('Type', axis=1)
y = df['Type']

# 1. Scale the features (Essential for KNN and SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Balance the classes using SMOTE
# This creates synthetic samples for minority classes to help reach 95%+ accuracy
smote = SMOTE(random_state=42, k_neighbors=3)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# 3. Split into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print("Data Preprocessing Complete.")

# Initialize and train KNN
knn_model = KNeighborsClassifier(n_neighbors=1, metric='manhattan', weights='uniform')
knn_model.fit(X_train, y_train)

# Predictions
knn_pred = knn_model.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)

print(f"--- KNN Performance ---")
print(f"Accuracy: {knn_acc:.2%}")
print("\nClassification Report:\n", classification_report(y_test, knn_pred))
```

```python
# 2) Implementing Linear SVM using scikit-learn
# - Use the same dataset (glass.csv)
# - Use train_test_split to create train/test
# - Evaluate using .score() and classification_report(y_true, y_pred)

#Trail 1

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report

# Load dataset
dst_Data = pd.read_csv("glass.csv")
dst_Data.info()  # (or use "/mnt/data/glass.csv" in Colab)

# Split into features (X) and target (y) -> assuming last column is target
X = dst_Data.iloc[:, :-1]
y = dst_Data.iloc[:, -1]

# Train-test split (stratify keeps class distribution similar in train/test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y
)

# Linear SVM works best with scaled features, so use a pipeline
svm_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", LinearSVC(random_state=0, max_iter=20000))
])

# Train the model
svm_model.fit(X_train, y_train)

# Predict on test set
y_pred = svm_model.predict(X_test)

# Evaluate: score() gives accuracy on test set
print("Accuracy (score):", svm_model.score(X_test, y_test) * 100)

# Detailed evaluation
print("Classification Report:\n", classification_report(y_test, y_pred))
```

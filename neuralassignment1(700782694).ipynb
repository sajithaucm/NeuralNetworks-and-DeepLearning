```python
#Q1(B)
#– Take two numbers from user and perform at least 4 arithmetic operations on them.
num1 = int(input("Enter First number"))
num2 = int(input("Enter Second number"))
print("Addition:", num1+num2)
print("Substraction:", num1-num2)
print("Multiplcation:", num1*num2)
if num2!=0 :
    print("Divsion:",num1/num2)
else:
    print("Division: Not Possible(Division by zero)")
```

```python
#Q3 Using the if statement conditions to write a program 
#print the letter grade based on an input class score. Use the grading scheme we are using in this class.
score = int(input("Enter Score:"))
if score >=90:
    print(" Grade (A)")
elif score >=80:
    print(" Grade (B)")
elif score >=70:
    print(" Grade (C)")
elif score >=60:
    print(" Grade (D)")
else:
    print("F")
```
